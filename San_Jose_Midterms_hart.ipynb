```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/heart_2020_cleaned.csv')

# Extract relevant columns
heart_disease = data['HeartDisease']
gender = data['Sex']

# Filter data by gender
male_data = data[data['Sex'] == 'Male']
female_data = data[data['Sex'] == 'Female']

# Count the occurrences of each category in the 'HeartDisease' column for each gender
male_heart_disease_counts = male_data['HeartDisease'].value_counts()
female_heart_disease_counts = female_data['HeartDisease'].value_counts()

# Plotting the pie chart for males
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plt.pie(male_heart_disease_counts, labels=male_heart_disease_counts.index, autopct='%1.1f%%', startangle=90, colors=['b', 'r'])
plt.title('Distribution of Heart Disease in Males')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.

# Plotting the pie chart for females
plt.subplot(1, 2, 2)
plt.pie(female_heart_disease_counts, labels=female_heart_disease_counts.index, autopct='%1.1f%%', startangle=90, colors=['r', 'b'])
plt.title('Distribution of Heart Disease in Females')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.

plt.show()
```

The second question Count the occurrences of each category in the 'HeartDisease' column for each gender.
