```python
import matplotlib.pyplot as plt #adding module
import numpy as np

xpoints = np.array([0, 6]) #plotting x and y points
ypoints = np.array([0, 10])

plt.plot(xpoints, ypoints, 'o') #plot is used to draw point
plt.show()
```

```python
#plotting x and y points
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker='x') #plot() is used to draw point
plt.show()
```

Format Strings fmt

```python
#plotting x and y poits
ypoints=np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:b') #marker | line | color
plt.show()
```

Marker Size

```python
#Plottinf x and y points
ypoints=np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:b', ms=20) #marker | line | color
plt.show()
```

Marker Color-You can use the argument markeredgecolor or mec for short

```python
#plotting x and y points
ypoints=np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:b', ms=20, mec='r') #marker | line | color
plt.show()
```

```python
#plotting x and y points
ypoints=np.array([3, 8, 1, 10])

plt.plot(ypoints, '*:b', ms=20, mec='r', mfc='red') #marker | line | color
plt.show()
```

Dashed

```python
plt.plot(ypoints,linestyle='dashed')
```

Matplot Labels and Title-You can use the xlabel() and ylabel()

```python
xpoints=np.array([1,2,6,8]) #plotting x and y points
ypoints=np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints) #plot() is used to draw points

plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km") #labels
plt.ylabel("speed in km/s")
plt.show()
```

Matplot grid

```python
xpoints = np.array([1, 2, 6, 8]) #plotting x and y points
ypoints = np.array([3, 8,1,10])

plt.plot(xpoints, ypoints) #plot() is used to draw points

plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km") #labels
plt.ylabel("speed in km/s")

plt.grid(axis='y')
plt.show()
#
```

```python
#select grid lines to display
xpoints = np.array([1, 2, 6, 8]) #plotting x and y points
ypoints = np.array([3, 8,1,10])

plt.plot(xpoints, ypoints) #plot() is used to draw points
plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km") #labels
plt.ylabel("speed in km/s")

plt.grid(axis='x', linestyle='--', linewidth=0.50)
plt.show()
```

Part 1: Perform the following codes, and understand the difference between line plot, scatter plot, histogram, bar chart, box plot, and pie chart using matplotlib’s pyplot sub-module. (Provide a screenshot of your output.)

```python
#Line Plot
import matplotlib.pyplot as plt

x = [1, 2, 3, 4] #plotting x and y points
y = [10, 20, 30, 40]
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#Scatter Plot
import matplotlib.pyplot as plt

x = [1, 2, 3, 4] #plotting x and y points
y = [10, 15, 35, 40]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#Histogram
import matplotlib.pyplot as plt

x = [1, 2, 2, 3, 3, 4, 4, 5]
plt.hist(x)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#Bar Chart
import matplotlib.pyplot as plt

x = ['A', 'B', 'C', 'D'] #plotting x and y points
y = [10, 20, 30, 40]
plt.bar(x, y)
plt.title("Bar Chart")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
#Box Plot
import matplotlib.pyplot as plt

# Sample data
data = [[1.5]*10 + [2]*10 + [3]*10]

# Create the boxplot
plt.boxplot(data)
# Add title and labels
plt.title("Box Plot")
plt.ylabel("Value")
# Display the plot
plt.show()
```

```python
#Pie Chart
import matplotlib.pyplot as plt
g = np.array([35,25,25,15])
grader=["Grade 1=35%", "Grade 2=25%", "Grade 3=25%", "Grade 4=15%"]
plt.pie(g, labels=grader)
plt.show()
```
