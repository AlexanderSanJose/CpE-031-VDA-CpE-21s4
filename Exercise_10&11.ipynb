```python

#Generate data
x = np.array([1,2,3,4,5])
y = np.array([2,4,5,4,5])

#Calculate the slope and y-intecept of the regression line
slope, intercept = np.polyfit(x,y,1)

#Calculate the predicted values of y using the regression line eqation
y_pred = slope*x + intercept

#Plot the original data points and the regression line
plt.scatter(x,y)
plt.plot(x, y_pred, color='red')
plt.show()
```

```python

x = np.array([60, 90, 70,85 ,80,65,75,60,75,70])
y = np.array([70,95, 70, 75, 90,75,75,60,80,65])

slope, intercept = np.polyfit(x, y, 1)

y_pred = slope * x + intercept

plt.scatter(x, y)
plt.plot(x, y_pred, color='red')


plt.title("Scatter Plot of correlation of exam score and course grade")
plt.xlabel("Exam Score")
plt.ylabel("Course Grade")
plt.show()
```
