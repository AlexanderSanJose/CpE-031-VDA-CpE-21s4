a. Read each device's date of release and sales. Show it using a line plot.

```python
path = pd.read_csv('/content/wTechnological-Products-Sample-Data - Tech Products.csv')
xpoints = path['Date of Release']#plotting x and y points
ypoints = path['Sales']
plt.figure(figsize=(10, 6))
plt.plot(xpoints, ypoints) #plot is used to draw point
plt.title("Price of Technological Products Over Time")
plt.xlabel("Date of Release")
plt.ylabel("Sales")
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
path = pd.read_csv('/content/wTechnological-Products-Sample-Data - Tech Products.csv')
xpoints = path['Date of Release']#plotting x and y points
ypoints = path['Sales']
plt.figure(figsize=(10, 6))
plt.title("Total Sales Over Time")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

plt.plot(xpoints, ypoints,  'o:r', linewidth=3, mfc='blue', ms=6, label='Sales Data' )
plt.legend(loc = 'lower right')
plt.show()
```

c. Read all product sales data and show it using a multiline plot. Label them.

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('/content/wTechnological-Products-Sample-Data - Tech Products.csv')


x = data['Date of Release']
y1 = data['Brand']
y2 = data['Device']
y3 = data['Model']

# Step 3: Create the plot
plt.figure(figsize=(10, 20))
plt.title("Product Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales (Units)")
plt.xticks(rotation=45)
plt.yticks(rotation=45)

plt.plot(x, y1, marker='o', color='r', label='Product A', linestyle='-')
plt.plot(x, y2, marker='s', color='g', label='Product B', linestyle='--')
plt.plot(x, y3, marker='^', color='b', label='Product C', linestyle=':')
plt.plot(label='Sales')

plt.legend(loc = 'upper left')
plt.tight_layout()
plt.show()
```

d. Read each brand sales data and show it using the bar chart. Label them.

```python


# Read the data
data = pd.read_csv('/content/wTechnological-Products-Sample-Data - Tech Products.csv')

brands = data['Brand']
sales = data['Sales']

# Plot
plt.figure(figsize=(10, 6))
plt.bar(brands, sales, color='skyblue')
plt.xlabel('Brand')
plt.ylabel('Sales')
plt.title('Sales Data by Brand')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

e. Read each device and show it using the histogram to see the most common Country of Origin. Label them.

```python

# Read the data
data = pd.read_csv('/content/wTechnological-Products-Sample-Data - Tech Products.csv')

country_of_origin = data['Country of Origin']

# Plot
plt.figure(figsize=(10, 6))
plt.hist(country_of_origin, bins=len(country_of_origin.unique()), color='skyblue', edgecolor='black')
plt.xlabel('Country of Origin')
plt.ylabel('Number of Devices')
plt.title('Most Common Country of Origin for Devices')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

f. Create a pie chart that shows each brand Country of Origin. Label them.

```python

# Read the data
data = pd.read_csv('/content/wTechnological-Products-Sample-Data - Tech Products.csv')

# Count the number of brands from each country of origin
country_counts = data['Country of Origin'].value_counts()

# Create pie chart
plt.figure(figsize=(10, 6))
wedges, texts, autotexts = plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=140, pctdistance=0.85)  # Adjust pctdistance for better readability

# Add brand information to the legend
legend_labels = [f"{country} ({', '.join(data[data['Country of Origin'] == country]['Brand'].unique())})" for country in country_counts.index]
plt.legend(wedges, legend_labels, title="Country of Origin (Brands)", loc="lower right", bbox_to_anchor=(0, 0.65))

plt.axis('equal')
plt.title('Brand Distribution by Country of Origin')
plt.tight_layout()
plt.show()
```
