# **5.) Selecting dataframe columns:**

```python
import pandas as pd

# Sample DataFrame
data = [
    {'Student Name': 'Alice', 'Quiz 1': 50, 'Python Fundamentals': 100, 'Conditional Statements': 100, 'CpE 0323-1': 1},
    {'Student Name': 'Bob', 'Quiz 1': 33, 'Python Fundamentals': 95, 'Conditional Statements': 90, 'CpE 0323-1': 2},
    {'Student Name': 'Charlie', 'Quiz 1': 22, 'Python Fundamentals': 90, 'Conditional Statements': 80, 'CpE 0323-1': 3},
    {'Student Name': 'David', 'Quiz 1': 42, 'Python Fundamentals': 85, 'Conditional Statements': 70, 'CpE 0323-1': 4},
    {'Student Name': 'Eve', 'Quiz 1': 34, 'Python Fundamentals': 80, 'Conditional Statements': 60, 'CpE 0323-1': 5},
]

grades = pd.DataFrame(data)

# Selecting a single column
single_column = grades['Quiz 1']
print("Single Column (Quiz 1):")
print(single_column)

# Selecting multiple columns
multiple_columns = grades[['Student Name', 'Quiz 1', 'Python Fundamentals']]
print('\n')
print("\nMultiple Columns (Student Name, Quiz 1, Python Fundamentals):")
print(multiple_columns)

grades
```

# **6.) Selecting dataframe rows:**

```python
import pandas as pd

# Sample DataFrame
data = [
    {'Student Name': 'Alice', 'Quiz 1': 50, 'Python Fundamentals': 100, 'Conditional Statements': 100, 'CpE 0323-1': 1},
    {'Student Name': 'Bob', 'Quiz 1': 33, 'Python Fundamentals': 95, 'Conditional Statements': 90, 'CpE 0323-1': 2},
    {'Student Name': 'Charlie', 'Quiz 1': 22, 'Python Fundamentals': 90, 'Conditional Statements': 80, 'CpE 0323-1': 3},
    {'Student Name': 'David', 'Quiz 1': 42, 'Python Fundamentals': 85, 'Conditional Statements': 70, 'CpE 0323-1': 4},
    {'Student Name': 'Eve', 'Quiz 1': 34, 'Python Fundamentals': 80, 'Conditional Statements': 60, 'CpE 0323-1': 5},
]

grades = pd.DataFrame(data)

# Selecting a single row using pandas indexing (iloc)
single_row_iloc = grades.iloc[2]
print("Single Row (iloc):")
print(single_row_iloc)

# Selecting multiple rows using pandas indexing (iloc)
multiple_rows_iloc = grades.iloc[1:4]
print("\nMultiple Rows (iloc):")
print(multiple_rows_iloc)

# Selecting a single row using Python indexing
single_row_python = grades.iloc[2]
print("\nSingle Row (Python indexing):")
print(single_row_python)

# Selecting multiple rows using Python indexing
multiple_rows_python = grades.iloc[1:4]
print("\nMultiple Rows (Python indexing):")
print(multiple_rows_python)

grades
```
